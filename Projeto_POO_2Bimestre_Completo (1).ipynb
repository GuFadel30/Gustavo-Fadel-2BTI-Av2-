## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**

- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 2
# Classe principal
class Receita:
    def __init__(self, titulo, ingredientes, tempo_preparo):
        self.titulo = titulo
        self.ingredientes = ingredientes
        self.tempo_preparo = tempo_preparo

    def exibir_dados(self):
        print(f"Receita: {self.titulo}")
        print(f"Ingredientes: {', '.join(self.ingredientes)}")
        print(f"Tempo de preparo: {self.tempo_preparo} minutos")
        print("-" * 40)

    # M√©todo de a√ß√£o 1
    def adicionar_ingrediente(self, ingrediente):
        self.ingredientes.append(ingrediente)
        print(f"Ingrediente '{ingrediente}' adicionado com sucesso!")

    # M√©todo de a√ß√£o 2
    def alterar_tempo_preparo(self, novo_tempo):
        self.tempo_preparo = novo_tempo
        print(f"Tempo de preparo alterado para {novo_tempo} minutos.")


# Subclasse herdada de Receita
class ReceitaSobremesa(Receita):
    def __init__(self, titulo, ingredientes, tempo_preparo, nivel_acucar):
        super().__init__(titulo, ingredientes, tempo_preparo)
        self.nivel_acucar = nivel_acucar  # novo atributo espec√≠fico da sobremesa

    # M√©todo exclusivo da subclasse
    def verificar_acucar(self):
        if self.nivel_acucar > 50:
            print("Aten√ß√£o: Sobremesa com alto teor de a√ß√∫car!")
        else:
            print("N√≠vel de a√ß√∫car moderado.")


# Criando um objeto da subclasse
sobremesa1 = ReceitaSobremesa(
    "Pudim de Leite",
    ["leite condensado", "leite", "ovos", "a√ß√∫car"],
    90,
    nivel_acucar=60
)

# Testando m√©todos
sobremesa1.exibir_dados()
sobremesa1.verificar_acucar()
sobremesa1.adicionar_ingrediente("baunilha")
sobremesa1.alterar_tempo_preparo(95)
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 3
# Classe principal com encapsulamento
class Receita:
    def __init__(self, titulo, ingredientes, tempo_preparo):
        self.titulo = titulo
        self.ingredientes = ingredientes
        self.__tempo_preparo = tempo_preparo  # atributo privado

    # M√©todo para exibir dados
    def exibir_dados(self):
        print(f"Receita: {self.titulo}")
        print(f"Ingredientes: {', '.join(self.ingredientes)}")
        print(f"Tempo de preparo: {self.__tempo_preparo} minutos")
        print("-" * 40)

    # Getter para tempo de preparo
    def get_tempo_preparo(self):
        return self.__tempo_preparo

    # Setter para tempo de preparo
    def set_tempo_preparo(self, novo_tempo):
        if novo_tempo > 0:
            self.__tempo_preparo = novo_tempo
            print(f"Tempo de preparo atualizado para {novo_tempo} minutos.")
        else:
            print("Erro: o tempo de preparo deve ser positivo.")

    # M√©todo com l√≥gica interna (abstra√ß√£o)
    def verificar_preparo_rapido(self):
        if self.__tempo_preparo <= 30:
            return "Receita r√°pida!"
        else:
            return "Receita mais demorada."


# Teste com objeto
receita3 = Receita(
    "Omelete Simples",
    ["ovos", "sal", "queijo", "cebola"],
    15
)

# Usando os m√©todos definidos
receita3.exibir_dados()
print(receita3.verificar_preparo_rapido())

receita3.set_tempo_preparo(25)
print(f"Tempo atualizado: {receita3.get_tempo_preparo()} minutos")
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**

- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 4
import sqlite3

# Classe principal
class Receita:
    def __init__(self, titulo, ingredientes, tempo_preparo):
        self.titulo = titulo
        self.ingredientes = ingredientes
        self.__tempo_preparo = tempo_preparo

    def get_tempo_preparo(self):
        return self.__tempo_preparo

    def set_tempo_preparo(self, tempo):
        if tempo > 0:
            self.__tempo_preparo = tempo

    # M√©todo que ser√° sobrescrito
    def exibir_dados(self):
        print(f"Receita: {self.titulo}")
        print(f"Ingredientes: {', '.join(self.ingredientes)}")
        print(f"Tempo de preparo: {self.__tempo_preparo} minutos")
        print("-" * 40)


# Subclasse com polimorfismo
class ReceitaSobremesa(Receita):
    def __init__(self, titulo, ingredientes, tempo_preparo, nivel_acucar):
        super().__init__(titulo, ingredientes, tempo_preparo)
        self.nivel_acucar = nivel_acucar

    # Polimorfismo: sobrescrevendo exibir_dados
    def exibir_dados(self):
        super().exibir_dados()
        print(f"N√≠vel de a√ß√∫car: {self.nivel_acucar}%")
        print("-" * 40)


# Banco de dados: conex√£o e cria√ß√£o da tabela
def criar_banco():
    conexao = sqlite3.connect("receitas.db")
    cursor = conexao.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS receitas (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            titulo TEXT,
            ingredientes TEXT,
            tempo_preparo INTEGER
        )
    ''')
    conexao.commit()
    conexao.close()


# Inserir receita no banco
def inserir_receita(titulo, ingredientes, tempo_preparo):
    conexao = sqlite3.connect("receitas.db")
    cursor = conexao.cursor()
    cursor.execute('''
        INSERT INTO receitas (titulo, ingredientes, tempo_preparo)
        VALUES (?, ?, ?)
    ''', (titulo, ', '.join(ingredientes), tempo_preparo))
    conexao.commit()
    conexao.close()
    print("Receita inserida com sucesso no banco de dados!")


# Consultar todas as receitas
def consultar_receitas():
    conexao = sqlite3.connect("receitas.db")
    cursor = conexao.cursor()
    cursor.execute("SELECT * FROM receitas")
    resultados = cursor.fetchall()
    conexao.close()
    print("\nüìã RECEITAS NO BANCO DE DADOS:")
    for linha in resultados:
        print(f"ID: {linha[0]}, T√≠tulo: {linha[1]}, Ingredientes: {linha[2]}, Tempo: {linha[3]} minutos")


# Excluir uma receita por ID
def deletar_receita(id_receita):
    conexao = sqlite3.connect("receitas.db")
    cursor = conexao.cursor()
    cursor.execute("DELETE FROM receitas WHERE id = ?", (id_receita,))
    conexao.commit()
    conexao.close()
    print(f"Receita com ID {id_receita} exclu√≠da com sucesso.")


# Execu√ß√£o de teste
if __name__ == "__main__":
    criar_banco()

    # Criando e exibindo uma receita comum
    receita_comum = Receita("Macarr√£o Alho e √ìleo", ["macarr√£o", "alho", "√≥leo", "sal"], 20)
    receita_comum.exibir_dados()
    inserir_receita(receita_comum.titulo, receita_comum.ingredientes, receita_comum.get_tempo_preparo())

    # Criando e exibindo uma sobremesa com polimorfismo
    sobremesa = ReceitaSobremesa("Mousse de Lim√£o", ["leite condensado", "creme de leite", "suco de lim√£o"], 15, 45)
    sobremesa.exibir_dados()
    inserir_receita(sobremesa.titulo, sobremesa.ingredientes, sobremesa.get_tempo_preparo())

    # Consultar receitas no banco
    consultar_receitas()

    # Exemplo de exclus√£o (remova o coment√°rio se quiser testar)
    # deletar_receita(1)  # Altere o ID conforme necess√°rio
```

**C√≥digo:**

```python
!pip install mysql-connector-python

import mysql.connector

# ================================
# CONFIGURA√á√ÉO DO BANCO DE DADOS
# ================================
def conectar_banco():
    return mysql.connector.connect(
        host="localhost",
        user="root",
        password="",  # ou a senha do seu MySQL
        database="site_receitas"
    )

# ================================
# CLASSE PRINCIPAL: Receita
# ================================
class Receita:
    def __init__(self, titulo, ingredientes, tempo_preparo):
        self.titulo = titulo
        self.ingredientes = ingredientes
        self.__tempo_preparo = tempo_preparo

    def get_tempo_preparo(self):
        return self.__tempo_preparo

    def set_tempo_preparo(self, tempo):
        if tempo > 0:
            self.__tempo_preparo = tempo
        else:
            print("Tempo inv√°lido!")

    def verificar_preparo_rapido(self):
        return self.__tempo_preparo <= 30

    def exibir_dados(self):
        print(f"\nReceita: {self.titulo}")
        print(f"Ingredientes: {', '.join(self.ingredientes)}")
        print(f"Tempo de preparo: {self.__tempo_preparo} minutos")
        print("R√°pida?", "Sim" if self.verificar_preparo_rapido() else "N√£o")
        print("-" * 40)

# ================================
# SUBCLASSE: ReceitaSobremesa
# ================================
class ReceitaSobremesa(Receita):
    def __init__(self, titulo, ingredientes, tempo_preparo, nivel_acucar):
        super().__init__(titulo, ingredientes, tempo_preparo)
        self.nivel_acucar = nivel_acucar

    def exibir_dados(self):
        super().exibir_dados()
        print(f"N√≠vel de a√ß√∫car: {self.nivel_acucar}%")
        if self.nivel_acucar > 50:
            print("‚ö†Ô∏è Aten√ß√£o: Alto teor de a√ß√∫car!")
        print("-" * 40)

# ================================
# FUN√á√ïES DE BANCO DE DADOS
# ================================
def inserir_receita(titulo, ingredientes, tempo_preparo, nivel_acucar=None):
    conn = conectar_banco()
    cursor = conn.cursor()

    try:
        # Inserir na tabela de receitas
        cursor.execute('''
            INSERT INTO receitas (titulo, ingredientes, tempo_preparo)
            VALUES (%s, %s, %s)
        ''', (titulo, ', '.join(ingredientes), tempo_preparo))
        conn.commit()
        receita_id = cursor.lastrowid

        # Se for sobremesa, insere na tabela de sobremesas
        if nivel_acucar is not None:
            cursor.execute('''
                INSERT INTO sobremesas (id_receita, nivel_acucar)
                VALUES (%s, %s)
            ''', (receita_id, nivel_acucar))
            conn.commit()

        print(f"‚úÖ Receita '{titulo}' adicionada ao banco de dados.")
    except Exception as e:
        print("Erro ao inserir no banco:", e)
        conn.rollback()
    finally:
        cursor.close()
        conn.close()

def consultar_receitas():
    conn = conectar_banco()
    cursor = conn.cursor()

    try:
        cursor.execute('''
            SELECT r.id, r.titulo, r.ingredientes, r.tempo_preparo, s.nivel_acucar
            FROM receitas r
            LEFT JOIN sobremesas s ON r.id = s.id_receita
        ''')
        receitas = cursor.fetchall()

        print("\nüìã RECEITAS CADASTRADAS NO BANCO:")
        for r in receitas:
            print(f"ID: {r[0]} | T√≠tulo: {r[1]} | Ingredientes: {r[2]} | Tempo: {r[3]} min", end='')
            if r[4] is not None:
                print(f" | A√ß√∫car: {r[4]}%")
            else:
                print()
    except Exception as e:
        print("Erro na consulta:", e)
    finally:
        cursor.close()
        conn.close()

def deletar_receita(id_receita):
    conn = conectar_banco()
    cursor = conn.cursor()

    try:
        cursor.execute("DELETE FROM receitas WHERE id = %s", (id_receita,))
        conn.commit()
        print(f"üóëÔ∏è Receita com ID {id_receita} foi exclu√≠da.")
    except Exception as e:
        print("Erro ao excluir receita:", e)
        conn.rollback()
    finally:
        cursor.close()
        conn.close()

# ================================
# ADICIONAR RECEITA
# ================================
def adicionar_receita():
    titulo = input("Digite o t√≠tulo da receita: ")
    ingredientes = input("Digite os ingredientes (separados por v√≠rgula): ").split(",")

    while True:
        try:
            tempo_preparo = int(input("Digite o tempo de preparo em minutos: "))
            break
        except ValueError:
            print("‚ö†Ô∏è Por favor, digite um n√∫mero inteiro v√°lido para o tempo!")

    sobremesa = input("√â uma sobremesa? (s/n): ").strip().lower()

    if sobremesa == 's':
        while True:
            try:
                nivel_acucar = int(input("Qual o n√≠vel de a√ß√∫car (em %)? "))
                break
            except ValueError:
                print("‚ö†Ô∏è Digite um n√∫mero v√°lido para o n√≠vel de a√ß√∫car!")
        
        receita = ReceitaSobremesa(titulo, ingredientes, tempo_preparo, nivel_acucar)
    else:
        receita = Receita(titulo, ingredientes, tempo_preparo)
        nivel_acucar = None

    receita.exibir_dados()
    inserir_receita(titulo, ingredientes, tempo_preparo, nivel_acucar)

# ================================
# MENU
# ================================
def menu():
    while True:
        print("\n==== MENU ====")
        print("1 - Adicionar nova receita")
        print("2 - Consultar receitas")
        print("3 - Deletar receita")
        print("4 - Sair")
        
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            adicionar_receita()
        elif opcao == "2":
            consultar_receitas()
        elif opcao == "3":
            try:
                id_receita = int(input("Digite o ID da receita a excluir: "))
                deletar_receita(id_receita)
            except ValueError:
                print("‚ö†Ô∏è Por favor, digite um n√∫mero inteiro v√°lido para o ID.")
        elif opcao == "4":
            print("Volte para mais receitas!")
            break
        else:
            print("‚ùå Op√ß√£o inv√°lida. Tente novamente.")

# ================================
# MAIN
# ================================
def main():
    print("‚úÖ Sistema de receitas iniciado com MySQL!")
    menu()

if __name__ == "__main__":
    main()
```
